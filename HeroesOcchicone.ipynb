```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
df = pd.read_csv(file_to_load)
df.head()
```

```python
#Player Count
#calculate player count and create a dataframe of the results

pcgdf = df.groupby(["SN"])
pcount = pcgdf["SN"].unique().count()
pcountdf = pd.DataFrame([{"Player Count": pcount}])
pcountdf
```

```python
#Purchasing Analysis Total
#calculate unique item count, average price, total revenue and total number of purchases and set as variables
group = df.groupby(["Item ID"])
icount = group["Item ID"].unique().count()

tprice = df["Price"].sum(axis=0)
pucount = df["Purchase ID"].count()
aprice = df["Price"].mean()

#create dataframe to display variables
purdf = pd.DataFrame([{"Number of Unique Items": icount, "Average Price" : aprice, "Number of Purchases" : pucount, 
                       "Total Revenue" : tprice}])

#format $$ correctly
purdf["Average Price"] = purdf["Average Price"].map("${:,.2f}".format)
purdf["Total Revenue"] = purdf["Total Revenue"].map("${:,.2f}".format)
purdf
```

```python
#Gender Demographics
#calculate unique males and %.
malesc = df.loc[df["Gender"] == "Male"]
malesgr = malesc.groupby(["SN"])
mcount = malesgr["SN"].unique().count()
pmales = (mcount/pcount)*100

#calculate unique females and %.
femalec = df.loc[df["Gender"] == "Female"]
femalegr = femalec.groupby(["SN"])
fcount = femalegr["SN"].unique().count()
pfemales = (fcount/pcount)*100

#calculate unique other and %.
otherc = df.loc[df["Gender"] == "Other / Non-Disclosed"]
othergr = otherc.groupby(["SN"])
ocount = othergr["SN"].unique().count()
pother = (ocount/pcount)*100

#create dataframe
genderdf = pd.DataFrame({"": ["Male", "Female", "Other / Non-Disclosed"],  "Total Count" : [mcount, fcount, ocount],  
                          "Percentage of Players" : [pmales, pfemales, pother]}) 

#format %
genderdf["Percentage of Players"] = genderdf["Percentage of Players"].map("{:,.2f}%".format)

genderdf
```

```python
#Purchasing Analysis by Gender

#group by gender
gender2 = df.groupby(["Gender"])

#calculate variables
genavepp = gender2["Price"].mean()
gentpv = gender2["Price"].sum()
genpc = gender2["Price"].count()

#create a list for counts of gender
ugen = [fcount, mcount, ocount]

#create dataframe
gender2df = pd.DataFrame({"Purchase Count" : genpc, "Average Purchase Price" : genavepp, "Total Purchase Value" : gentpv}) 
gender2df["Avg Total Purchases Per Person"] = gender2df["Total Purchase Value"]/ugen

#formate $$
gender2df["Average Purchase Price"] = gender2df["Average Purchase Price"].map('${:,.2f}'.format)
gender2df["Total Purchase Value"] = gender2df["Total Purchase Value"].map('${:,.2f}'.format)
gender2df["Avg Total Purchases Per Person"] = gender2df["Avg Total Purchases Per Person"].map('${:,.2f}'.format)
gender2df
```

```python
#Age Demographics
#list unique players
dfunique = df.drop_duplicates("SN", keep='first', inplace=False)

#create bins and bin names
bins = [0, 9, 14, 19, 24, 29, 34, 39, 46]
groupnames = ["<10", "10-14", "15-19", "20-24", "25-30", "31-34", "35-39", "40+"]
dfunique["Age Category"] = pd.cut(dfunique["Age"], bins, labels = groupnames, include_lowest=True)
dfunique["Age Category"].value_counts()

#group data by Age Category 
agegroups = dfunique.groupby(["Age Category"])

#calculate count and %
age_count = agegroups["SN"].count()
age_perc = ((age_count/pcount)*100)

#create dataframe
agesumdf = pd.DataFrame({"Total Count" : age_count, "Percentage of Players" : age_perc})

#format %
agesumdf["Percentage of Players"] = agesumdf["Percentage of Players"].map("{:,.2f}%".format)
agesumdf
```

```python
#Purchasing Analysis by Age
#create bins and group names
bins = [0, 9, 14, 19, 24, 29, 34, 39, 46]
groupnames = ["<10", "10-14", "15-19", "20-24", "25-30", "31-34", "35-39", "40+"]
df["Age Ranges"] = pd.cut(df["Age"], bins, labels = groupnames, include_lowest=True)

#group by Age Range
ab_df = df.groupby(["Age Ranges"])

#calcualte variables
ab_count = ab_df["SN"].count()
ab_mean = ab_df["Price"].mean()
ab_tvalue = ab_df["Price"].sum()
ab_meanpp = (ab_tvalue/age_count)

#create dataframe
ab_summary = pd.DataFrame({"Purchase Count" : ab_count,"Average Purchase Price" : ab_mean, 
                          "Total Purchase Value" : ab_tvalue, "Ave Total Purchase Per Person" : ab_meanpp})

#format dataframe
ab_summary["Average Purchase Price"] = ab_summary["Average Purchase Price"].map('${:,.2f}'.format)
ab_summary["Total Purchase Value"] = ab_summary["Total Purchase Value"].map('${:,.2f}'.format)
ab_summary["Ave Total Purchase Per Person"] = ab_summary["Ave Total Purchase Per Person"].map('${:,.2f}'.format)
ab_summary
```

```python
#Top Spenders
#calculate purchase count
purchase_count = df["SN"].value_counts()

#calculate total spend and average spend
groupeddf = df.groupby(["SN"])
groupeddf.count()
total_spend = groupeddf["Price"].sum()
ave_spend = groupeddf["Price"].mean()

#create and sort dataframe
summarydf = pd.DataFrame({"Purchase Count" : purchase_count,"Average Purchase Price" : ave_spend, 
                          "Total Purchase Value" : total_spend})
summarydfsort = summarydf.sort_values(["Total Purchase Value"], ascending=False)

#format to $$
summarydfsort["Average Purchase Price"] = summarydfsort["Average Purchase Price"].map('${:,.2f}'.format)
summarydfsort["Total Purchase Value"] = summarydfsort["Total Purchase Value"].map('${:,.2f}'.format)

summarydfsort.head()
```

```python
#Most Popular Items
#Filter for requested columns
item_df = df[["Item ID", "Item Name", "Price"]]

#group by item ID and item name
item_gdf = item_df.groupby(["Item ID", "Item Name"])

#calculate count and total price
item_count = item_gdf.count()
item_tprice = item_gdf.sum()

#create data frame and calculate Item Price and rename columns
item_sdf= pd.DataFrame(item_gdf["Price"].sum())
item_sdf["Purchase Count"] = item_sdf["Purchase Count"] = item_count
item_sdf["Item Price"] = item_sdf["Item Price"] = (item_tprice/item_count)

#rename and reorganize columns
item_sdf = item_sdf.rename(columns={"Price":"Total Purchase Value"})
item_sdf = item_sdf[["Purchase Count", "Item Price", "Total Purchase Value"]]

#format to $$ and sort by Purchase Count
item_sdf1 = item_sdf.sort_values(["Purchase Count"], ascending=False)
item_sdf1["Item Price"] = item_sdf1["Item Price"].map('${:,.2f}'.format)
item_sdf1["Total Purchase Value"] = item_sdf1["Total Purchase Value"].map('${:,.2f}'.format)

item_sdf1.head()
```
